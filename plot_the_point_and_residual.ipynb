```python
import numpy as np
import matplotlib.pyplot as plt


z = np.load(r"/content/drive/MyDrive/Colab Notebooks/residual (2).npz")
points=z['points']
pointsx=points[:,:,0]
pointsy=points[:,:,1]
#print(points[0,99,0])
#print(points[0,99,1])
#print(points[1,0,0])
#print('points =', points)
#print('pointsx =', pointsx)
#print('pointsy =', pointsy)
#print(points.shape)
#print(pointsx.shape)
#print(pointsy.shape)



R=z['R'].reshape(-1,100,100)


gridx,gridy=np.meshgrid(np.linspace(-1,1,100),np.linspace(-1,1,100))


N=R.shape[0]
for i in range(N):
  plt.contourf(gridx,gridy,R[i])
  plt.scatter(pointsx[i],pointsy[i],color='r')
  plt.savefig('/content/figure1/Cylinder wake t=%.1f.jpg' % i)
  plt.close()


import imageio
images = []
for i in range(N):
    filename='/content/figure1/Cylinder wake t=%.1f.jpg' % i
    images.append(imageio.imread(filename))
imageio.mimsave('movie.gif', images)
```
